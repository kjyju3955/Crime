```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from sklearn.metrics import mean_squared_error
import math
import statsmodels.api as sm
from patsy import dmatrices
from statsmodels.stats.outliers_influence import variance_inflation_factor
from scipy.optimize import minimize
from scipy import sparse
import statsmodels.base.model as base
from statsmodels.iolib import summary2
from statsmodels.genmod import families
import warnings
import statsmodels
import patsy
%matplotlib inline

Data=pd.read_csv("./data/Test.csv",encoding='utf-8')
all_data=Data[Data['A']=='v']
all_data=all_data.filter(['State','Police_area','GDP','Penalty_mean','GDP_penalty','PCPI','Education','Happy','Misdemeanor_Result','MSA_Rate','Traveler_spending','older_rate','Female_Rate','Temper'])
all_data=all_data.reset_index()
#all_data
```

# All

```python
all_data_drop=all_data.drop(['Misdemeanor_Result'],axis=1)
all_target=all_data[['Misdemeanor_Result']]
all_x_data=all_data[['GDP','Temper','Police_area','GDP_penalty','Education','Happy','PCPI','MSA_Rate','Traveler_spending','older_rate','Female_Rate']]

all_x_data=sm.add_constant(all_x_data, prepend=False)
all_glm_binom = sm.GLM(all_target, all_x_data)
res = all_glm_binom.fit()
print(res.summary())
```
